```python
%matplotlib inline
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(20.0,10.0)

df=pd.read_csv(r'C:\Users\George.Macharia\Desktop\mlfiles\data2.csv')
```

```python
X=df['Age'].values
Y=df['Potential'].values
```

```python
mean_x=np.mean(X)
mean_y=np.mean(Y)

n=len(X)
num=0
den=0
for i in range(n):
    num+=(X[i]-mean_x)*(Y[i]-mean_y)
    den+=(X[i]-mean_x)**2
#getting m
t=num/den
#getting value of c
c=mean_y-(t*mean_x)
print(t,c,n)
```
